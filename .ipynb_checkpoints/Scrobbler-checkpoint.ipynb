```python
from datetime import datetime
```

```python
class Scrobbler:
    def __init__(self):  
        self.playlist = {}
    
    def print_feed(self, user):
        print("User:", user)
        
        print("1. Show Movie feed")
        print("2. Show Music feed")
        print("3. Show Podcast feed")
        
        menu = int(input("Choose Action: "))
        if menu == 1:
            print(self.playlist[user]['Movie'])
        elif menu == 2:
            print(self.playlist[user]['Song'])
        else:    
            print(self.playlist[user]['Podcast'])
    
    def run(self, user, content):
        now = datetime.now()
        timestamp = now.strftime("%m/%d/%Y, %H:%M:%S")
        
        setattr(content, "timestamp", timestamp)
        
        if user not in self.playlist.keys():
            self.playlist[user] = {}
            self.playlist[user]['Movie'] = []
            self.playlist[user]['Song'] = []
            self.playlist[user]['Podcast'] = []
            
            if type(content).__name__ == 'Song':
                self.playlist[user]['Song'] = [content]
                print('Add Song')
                
            elif type(content).__name__ == 'Movie':
                self.playlist[user]['Movie'] = [content]
                print('Add Movie')
                
            elif type(content).__name__ == 'Podcast':
                self.playlist[user]['Podcast'] = [content]
                print('Add Podcast')
        else:
            if type(content).__name__ == 'Song':
                self.playlist[user]['Song'].append(content)
                print('Add Song')
                print(self.playlist[user]['Song'])
            elif type(content).__name__ == 'Movie':
                self.playlist[user]['Movie'].append(content)
                print('Add Movie')
            elif type(content).__name__ == 'Podcast':
                self.playlist[user]['Podcast'].append(content)
                print('Add Podcast')
            
        
```

```python
class Song: 
    def __init__(self, title, artist, album, timestamp=None):
        self.title = title
        self.artist = artist
        self.album = album
        self.timestamp = timestamp
    
    def __str__(self):
        return f'Song: ({self.title} - {self.artist} / {self.album})' f'Played on: {self.timestamp}'
    
    def __repr__(self):
        return (f'<Song(title="{self.title}", artist="{self.artist}", album="{self.album}",timestamp="{self.timestamp}")>')

    
class Movie: 
    def __init__(self, title, year, timestamp=None):
        self.title = title
        self.year = year
        self.timestamp = timestamp
        
    def __str__(self):
        return f'Movie: {self.title} ({self.year})' 
    
    def __repr_(self):
        return (f'<Movie(title="{self.title}", year="{self.year}", timestamp="{self.timestamp}")>')
        
class Podcast: 
    def __init__(self, name, episode, timestamp=None):
        self.name = name
        self.episode = episode
        self.timestamp = timestamp      
    def __str__(self):
        return f'Podcast: {self.name} ({self.episode})' 
    
    def __repr_(self):
        return (f'<Podcast(name="{self.name}", episode="{self.episode}", timestamp="{self.timestamp}")>')
```
